```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully!")
```

```python
# Load the datasets

resume_df = pd.read_csv("../data/raw/Resume_data.csv")
linkedin_df = pd.read_csv("../data/raw/Linkedin_dataset.csv")
naukri_df = pd.read_csv("../data/raw/naukri.csv")

print("Datasets loaded successfully!")
```

```python
print("Resume Dataset")
display(resume_df.head())

print("LinkedIn Dataset")
display(linkedin_df.head())

print("Naukri Dataset")
display(naukri_df.head())
```

```python
print("Resume Dataset Shape:", resume_df.shape)
print("LinkedIn Dataset Shape:", linkedin_df.shape)
print("Naukri Dataset Shape:", naukri_df.shape)
```

```python
print("Resume Columns")
print(resume_df.columns)

print("\nLinkedIn Columns")
print(linkedin_df.columns)

print("\nNaukri Columns")
print(naukri_df.columns)
```

```python
print("========== Resume Dataset ==========")
resume_df.info()

print("\n========== LinkedIn Dataset ==========")
linkedin_df.info()

print("\n========== Naukri Dataset ==========")
naukri_df.info()
```

```python
print("Resume Dataset")
print(resume_df.isnull().sum())

print("\nLinkedIn Dataset")
print(linkedin_df.isnull().sum())

print("\nNaukri Dataset")
print(naukri_df.isnull().sum())
```

```python
print("Resume Dataset Duplicates:", resume_df.duplicated().sum())
print("LinkedIn Dataset Duplicates:", linkedin_df.duplicated().sum())
print("Naukri Dataset Duplicates:", naukri_df.duplicated().sum())
```

```python
# Remove duplicate rows from Resume dataset

resume_df = resume_df.drop_duplicates()

print("Resume Dataset Shape After Removing Duplicates:")
print(resume_df.shape)
```

```python
# Fill missing values in skills column

naukri_df["skills"] = naukri_df["skills"].fillna("Not Specified")

print("Missing values after filling:")
print(naukri_df.isnull().sum())
```

```python
print("Resume Dataset Shape:", resume_df.shape)
print("LinkedIn Dataset Shape:", linkedin_df.shape)
print("Naukri Dataset Shape:", naukri_df.shape)

print("\nMissing Values in Naukri Dataset:")
print(naukri_df.isnull().sum())
```

Exploratory data Analysis(EDA)

```python
#Resume category Distribution
plt.figure(figsize=(10,6))

resume_df["Category"].value_counts().plot(kind="bar")

plt.title("Resume Categories")
plt.xlabel("Category")
plt.ylabel("Number of Resumes")
plt.xticks(rotation=90)

plt.show()
```

```python
#Top jobs titles in linkdin dataset
plt.figure(figsize=(12,6))

linkedin_df["title"].value_counts().head(15).plot(kind="bar")

plt.title("Top 15 LinkedIn Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

```python
#Top job titles in naukri datasets
plt.figure(figsize=(12,6))

naukri_df["title"].value_counts().head(15).plot(kind="bar")

plt.title("Top 15 Naukri Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

naukri_df["location"].value_counts().head(15).plot(kind="bar")

plt.title("Top Hiring Locations")
plt.xlabel("Location")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(12,6))

naukri_df["experience"].value_counts().head(15).plot(kind="bar")

plt.title("Experience Required")
plt.xlabel("Experience")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(14,6))

naukri_df["salary"].value_counts().head(15).plot(kind="bar", color="green")

plt.title("Top Salary Ranges")
plt.xlabel("Salary")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=90)

plt.show()
```

```python
skills = naukri_df["skills"].str.split(",")

skills = skills.explode()

skills = skills.str.strip()
```

```python
plt.figure(figsize=(14,6))

skills.value_counts().head(20).plot(kind="bar", color="orange")

plt.title("Top 20 Required Skills")
plt.xlabel("Skills")
plt.ylabel("Frequency")
plt.xticks(rotation=45)

plt.show()
```

```python
resume_df["Resume_Length"] = resume_df["Resume"].str.len()
```

```python
plt.figure(figsize=(10,6))

plt.hist(resume_df["Resume_Length"], bins=30)

plt.title("Resume Length Distribution")
plt.xlabel("Number of Characters")
plt.ylabel("Frequency")

plt.show()
```

```python
linkedin_df["Description_Length"] = linkedin_df["description"].str.len()
```

```python
plt.figure(figsize=(10,6))

plt.hist(linkedin_df["Description_Length"], bins=30)

plt.title("LinkedIn Job Description Length")
plt.xlabel("Number of Characters")
plt.ylabel("Frequency")

plt.show()
```
